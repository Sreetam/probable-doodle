```python
import requests 
from bs4 import BeautifulSoup
import pandas as pd
import numpy as np
```

```python
def splice(a, b, c):
  return a[(a.index(b) + len(b)) : a.index(c)]
```

```python
all_models = []
all_links = []
url = []
modellist = []
videos = []
basestr = 'https://www.stripparadise.com/GameExtGenMb.php?GNum='
vidheader = 'https://www.pornstripgames.com/HTMLGames/sourceH5/'
#range(570, 875)
for i in range(570, 875):
  r = requests.get(basestr + str(i))
  soup = BeautifulSoup(r.content, 'html.parser')
  try:
    game = str(soup.find_all('iframe')[0])
    game = splice(game, 'src="', '" style')
    if 'pornstripgames' not in game: continue
    #extracting model names
    models = min([str(i) for i in soup.find_all('div') if 'models' in str(i)], key=len)
    model_list_raw = splice(models, 'models:\n', '\n</div>')
    model_list = [splice(i, '<span>', '</span') for i in model_list_raw.split(',\n')]
    #getting video list
    r = requests.get(game)
    soup = BeautifulSoup(r.content, 'html.parser')
    vidlist = [splice(str(i), 'src="', '" type') for i in soup.find_all('video')]
    vidlist_array = [vidheader + i[i.rfind('/')+1:] for i in vidlist]
    #appending the data
    modellist.append(', '.join(model_list))
    all_models = all_models + model_list
    all_links = all_links + vidlist_array
    videos.append(vidlist_array)
    url.append(game)
  except:
    continue
df = pd.DataFrame()
df['game'] = url
df['models'] = modellist
df['videos'] = videos
all_models = list(set(all_models))
```

```python
dfh = pd.DataFrame()
dfh['game'] = [splice(i, 'HTMLGames/', '.html') for i in df['game'].values]
dfh['game'] = list('<a href=' + df['game'] + '>' + dfh['game'] + '</a>')
dfh['models'] = modellist
dfh['videos'] = [
 ', '.join(['<a href=' + i + '>' + i[(i.index('sourceH5/') + len('sourceH5/')) :] + '</a>' for i in ll]) for ll in df['videos'].values
 ]
```

```python
df.to_csv("raw_list.csv")
df.to_excel("raw_list.xlsx")
df.to_html("raw_list.html")
dfh.to_html('list_page.html', escape=False)
pd.DataFrame(all_links).to_csv("linklist.txt", header=None, index=None)
pd.DataFrame(all_models).to_csv("modellist.txt", header=None, index=None)
pd.DataFrame(all_models).to_html("modellist.html", header=None)
```
